```python
import pandas as pd
import requests
from time import sleep
```

```python
url_root = 'https://pokeapi.co/api/v2/pokemon/'
```

```python
id = 1
error_count = 0
pokedex = []
while error_count < 5:
    
    url = url_root + str(id)
    res = requests.get(url)
    if res.status_code == 200:
        data = res.json()
        
        # Extract data
        pokemon = {}
        pokemon['poke-name'] = data['name']
        pokemon['pokenumber'] = data['id']
        pokemon['height'] = data['height']
        pokemon['weight'] = data['weight']
        
        # Extract stats
        for k in data['stats']:
            pokemon[k['stat']['name']] = k['base_stat']

        print(f"{pokemon['poke-name']} added successfully")
        id += 1
        pokedex.append(pokemon)
        sleep(1)
    
    else:
        print(f'Pokemon {id} is missing from API')
        id += 1
        error_count += 1
        sleep(1)
```

```python
pokedex = pd.DataFrame(pokedex)
pokedex.rename(columns={'speed': 'speed_TARGET'}, inplace=True)
pokedex.head()
```

```python
pokedex.to_csv('./data/pokedex.csv', index=False)
```
